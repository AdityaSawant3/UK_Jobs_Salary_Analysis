```python
import re
import numpy as np

def clean_salary(s):
    if pd.isnull(s):
        return np.nan

    s = str(s).lower().replace('£', '').replace(',', '').strip()

    # Handle range like "£35,392.00 to £42,618.00 per year"
    range_match = re.match(r'([\d\.]+)\s*to\s*([\d\.]+)', s)
    if range_match:
        low = float(range_match.group(1))
        high = float(range_match.group(2))
        return (low + high) / 2  # take average of range

    # Handle hourly rates
    if 'per hour' in s or '/hour' in s:
        try:
            hourly_rate = float(re.sub(r'[^\d\.]', '', s))
            return hourly_rate * 40 * 52  # full-time conversion
        except:
            return np.nan

    # Handle yearly salaries
    if 'per year' in s or '/year' in s:
        try:
            return float(re.sub(r'[^\d\.]', '', s))
        except:
            return np.nan

    # Default fallback for just numbers
    try:
        return float(re.sub(r'[^\d\.]', '', s))
    except:
        return np.nan
```

```python
df['salary'] = df['salary'].apply(clean_salary)
```
